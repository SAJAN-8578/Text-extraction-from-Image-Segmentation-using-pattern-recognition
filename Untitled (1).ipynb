```python
testing_letter = pd.read_csv(r'C:\Users\sajan\.cache\kagglehub\datasets\crawford\emnist\versions\3\emnist-letters-test.csv')
training_letter = pd.read_csv(r'C:\Users\sajan\.cache\kagglehub\datasets\crawford\emnist\versions\3\emnist-letters-train.csv')
```

```python
print(training_letter.shape)
print(testing_letter.shape)
```

```python
y1 = np.array(training_letter.iloc[:,0].values)  # Labels
x1 = np.array(training_letter.iloc[:,1:].values) # Images
y2 = np.array(testing_letter.iloc[:,0].values)   # Labels for test set
x2 = np.array(testing_letter.iloc[:,1:].values)  # Images for test set
```

```python
import matplotlib.pylab as plt
fig,axes = plt.subplots(3,5,figsize=(10,8))
for i, ax in enumerate(axes.flat):
    ax.imshow(x1[i].reshape([28,28]))
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Flatten images for SVM (28x28 -> 784 pixels)
train_images_flat = x1 / 255.0
test_images_flat = x2 / 255.0

# No need to reshape to 4D, keep it as 2D (samples, features)
X_train_flat = train_images_flat
X_test_flat = test_images_flat

# Split data into train and validation sets (using raw labels y1, not one-hot encoded)
X_train, X_val, y_train, y_val = train_test_split(X_train_flat, y1, test_size=0.2, random_state=42)
```

```python
# Display some rows of the preprocessed data
print("First 5 rows of standardized features (X_train):")
print(X_train[:5])
print("First 5 corresponding target values (y_train):")
print(y_train[:5])
```

```python
# Train SVM model
svm_classifier = svm.SVC(kernel='rbf', verbose=False)
svm_classifier.fit(X_train, y_train)

# Evaluate on validation set
val_preds = svm_classifier.predict(X_val)
val_accuracy = accuracy_score(y_val, val_preds)
print(f"Validation Accuracy: {val_accuracy * 100:.2f}%")

# Test set evaluation
test_preds = svm_classifier.predict(X_test_flat)
test_accuracy = accuracy_score(y2, test_preds)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
# Necessary imports
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, classification_report, confusion_matrix
from sklearn.metrics import matthews_corrcoef
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate accuracy, precision, recall, F1 score, Matthew's Correlation
accuracy_svm = accuracy_score(y_val, val_preds)
precision_svm, recall_svm, f1_svm, _ = precision_recall_fscore_support(y_val, val_preds, average='weighted')

# Print the evaluation metrics
print("Accuracy:", accuracy_svm)
print("Precision:", precision_svm)
print("Recall:", recall_svm)
print("F1 Score:", f1_svm)
print("\nMatthew's Correlation Coefficient: ")
print(matthews_corrcoef(y_val, val_preds))
```

```python

# Confusion Matrix
conf_matrix = confusion_matrix(y_val, val_preds)
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix for SVM")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Matthews correlation coefficient
mcc = matthews_corrcoef(y_val, val_preds)
print("Matthews Correlation Coefficient (MCC):", mcc)

# Classification report
print("\nClassification Report:\n", classification_report(y_val, val_preds))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, matthews_corrcoef
import numpy as np
import pandas as pd

# Assuming you already have val_preds (predicted labels) and y_val (true labels)

# Generate classification report as a dictionary
class_report = classification_report(y_val, val_preds, output_dict=True)

# Convert the classification report into a pandas DataFrame for easier plotting
report_df = pd.DataFrame(class_report).transpose()

# Drop the 'accuracy' row as we will handle it separately
report_df = report_df.drop(['accuracy'], axis=0)

# Plot precision, recall, f1-score, and support for each class
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot precision
sns.barplot(x=report_df.index[:-1], y=report_df['precision'][:-1], ax=axes[0, 0], palette="Blues_d")
axes[0, 0].set_title('Precision for Each Class')
axes[0, 0].set_xlabel('Class')
axes[0, 0].set_ylabel('Precision')

# Plot recall
sns.barplot(x=report_df.index[:-1], y=report_df['recall'][:-1], ax=axes[0, 1], palette="Greens_d")
axes[0, 1].set_title('Recall for Each Class')
axes[0, 1].set_xlabel('Class')
axes[0, 1].set_ylabel('Recall')

# Plot F1-score
sns.barplot(x=report_df.index[:-1], y=report_df['f1-score'][:-1], ax=axes[1, 0], palette="Oranges_d")
axes[1, 0].set_title('F1-Score for Each Class')
axes[1, 0].set_xlabel('Class')
axes[1, 0].set_ylabel('F1-Score')

# Plot support
sns.barplot(x=report_df.index[:-1], y=report_df['support'][:-1], ax=axes[1, 1], palette="Purples_d")
axes[1, 1].set_title('Support for Each Class')
axes[1, 1].set_xlabel('Class')
axes[1, 1].set_ylabel('Support')

# Adjust layout for better readability
plt.tight_layout()
plt.show()

# Matthews correlation coefficient (MCC)
mcc = matthews_corrcoef(y_val, val_preds)
print("Matthews Correlation Coefficient (MCC):", mcc)
```
